# Handling missing values

```python
df.isnull().sum()
```

```python
df['mileage(kmpl)'].unique()
```

```python
# df['mileage(kmpl)'] = df['mileage(kmpl)'].astype(str)

# df['mileage(kmpl)'] = df['mileage(kmpl)'].str.extract(r'(\d+\.?\d*)')

# df['mileage(kmpl)'] = pd.to_numeric(df['mileage(kmpl)'], errors='coerce')
```

```python
df['mileage(kmpl)'] = df['mileage(kmpl)'].fillna(df['mileage(kmpl)'].median())
df['engine(cc)'] = df['engine(cc)'].fillna(df['engine(cc)'].median())
df['max_power(bhp)'] = df['max_power(bhp)'].fillna(df['max_power(bhp)'].median())
df['torque(Nm)'] = df['torque(Nm)'].fillna(df['torque(Nm)'].median())

## replacing the null values of registration_year with manufacturing_year
df['registration_year'] = df['registration_year'].fillna(df['manufacturing_year'])
```
