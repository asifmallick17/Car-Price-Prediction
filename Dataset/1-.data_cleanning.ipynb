```python
import pandas as pd
df=pd.read_csv("Used Car Dataset.csv")
df.head()
```

```python
df.reset_index(drop=True)
df.drop('Unnamed: 0', axis=1)
```

```python
df['registration_year'].unique()
```

```python
df.dtypes
```

```python
df['car_name']=df['car_name'].str.replace(r'^\d{4}\s*','',regex=True)
df['registration_year'] = df['registration_year'].astype(str)
df['registration_year'] = df['registration_year'].str.split('-').str[1]
df['engine(cc)'] = df['engine(cc)'].astype(str)
df['engine(cc)']=df['engine(cc)'].str.split('.').str[0]
df['mileage(kmpl)'] = df['mileage(kmpl)'].astype(str)
df['mileage(kmpl)']=df['mileage(kmpl)'].str.split('.').str[0]

df
```

```python
df['kms_driven'].unique
```

```python
df.reset_index(drop=True, inplace=True)

df.drop('Unnamed: 0', axis=1, inplace=True, errors='ignore')

df.head()
```

```python
## taking 1st 2 string from car_name column
df['car_name']=df['car_name'].str.split().str[0:2].str.join(' ')
df.head()
```

```python
df.isnull().sum()
```

```python
df.to_csv("Cleaned_data.csv",index=False)
```
